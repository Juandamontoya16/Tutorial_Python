# Módulos e importaciones

#### En Python, se pueden importar paquetes para extender su funcionalidad. Un paquete es una colección de módulos que proporciona funcionalidades específicas. Un módulo es un archivo de Python que contiene definiciones y declaraciones que pueden ser reutilizadas en otros programas.

### Diferentes formas de importar módulos en Python

```python
# 1. Importar todo un módulo
import math  # Importa el módulo math completo
print("Raíz cuadrada de 16:", math.sqrt(16))
```

```python
# 2. Importar una función específica de un módulo
from math import pi  # Importa solo la constante pi
print("El valor de pi es:", pi)
```

```python
# 3. Importar un módulo con un alias
import numpy as np  # Importa numpy con un alias más corto
array = np.array([1, 2, 3])
print("Array con numpy:", array)
```

```python
# 4. Importar todas las funciones de un módulo (no recomendado)
from math import *  # No es recomendable, puede generar conflictos de nombres
print("Exponencial de 2:", exp(2))
```

### Creación de módulos propios

#### Se puede crear un archivo llamado "mimodulo.py" con funciones propias e importarlo en otro script
#### Ejemplo:

```python
# """
def bienvenida(usuario):
    return f"Bienvenido, {usuario}!"
# """
# Luego, en otro archivo, se puede importar con: `from mimodulo import bienvenida`
```

## Ejemplo con Matplotlib

```python
import matplotlib.pyplot as plt  # Para generar gráficos

# matplotlib.pyplot: Es una biblioteca de trazado que permite generar gráficos de líneas, barras, dispersión y más.
```

```python
# Gráficas con Matplotlib
# Definir datos para el gráfico
tiempo = [1, 2, 3, 4, 5]  # Lista con valores de tiempo
velocidad = [5, 10, 20, 30, 45]  # Lista con valores de velocidad

# Crear la gráfica de línea con etiquetas y formato
plt.plot(tiempo, velocidad, marker='s', linestyle='-', color='r', label='Velocidad vs Tiempo')

# Agregar título y etiquetas a los ejes
plt.title("Gráfico de Velocidad en Python")  # Título del gráfico
plt.xlabel("Tiempo (s)")  # Etiqueta del eje X
plt.ylabel("Velocidad (m/s)")  # Etiqueta del eje Y

# Agregar leyenda y cuadrícula
plt.legend()  # Muestra la leyenda con la etiqueta definida en plot()
plt.grid(True)  # Activa la cuadrícula en el gráfico

# Mostrar la gráfica
plt.show()
```
